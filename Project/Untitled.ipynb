```python
y = np.array(df['Salary']).reshape(-1,1)
y
```

```python
x.size
```

```python
Sxy = (sum(x*y)-(sum(x)*sum(y))/x.size)
Sxy
```

```python
Sxx = (sum(np.square(x)) - np.square(sum(x))/x.size)
Sxx
```

```python
b1 = Sxy / Sxx
b1
```

```python
np.average(x)
```

```python
np.average(y)
```

```python
b0 = np.average(y) - b1*np.average(x)
b0
```
